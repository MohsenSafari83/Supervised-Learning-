## NumPy Array
Numpy array is a powerful N-dimensional array object which is in the form of rows and columns. We can initialize numpy arrays from nested Python lists and access it elements. In order to perform these numpy operations, the next question which will come in your mind is:


## What’s the difference between a Python list and a NumPy array?

NumPy gives you an enormous range of fast and efficient ways of creating arrays and manipulating numerical data inside them. While a Python list can contain different data types within a single list, all of the elements in a NumPy array should be homogeneous. The mathematical operations that are meant to be performed on arrays would be extremely inefficient if the arrays weren’t homogeneous.

## Why use NumPy?

NumPy arrays are faster and more compact than Python lists. An array consumes less memory and is convenient to use. NumPy uses much less memory to store data and it provides a mechanism of specifying the data types. This allows the code to be optimized even further.


##Data Management Basics

In Machine Learning, NumPy is widely used for handling and transforming data.  
Here are some of the most important functions:

- **Array attributes**: `shape`, `dtype`  
- **Reshaping data**: `reshape`, `arange`  
- **Managing elements**: adding, removing, sorting  
- **Flattening arrays**: `flatten()` vs. `ravel()`  


```python
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])

print(a.shape)   # (2, 3) → 2 rows, 3 columns
print(a.dtype)   # int64 (data type of elements)
```

```python
b = np.arange(12)
print(b)
```

```python
c = b.reshape(3, 4)         # reshape into 3 rows, 4 cols
print(c)
```

```python
print(np.reshape(a, (1, -1)))  # -1 means the number of columns will be determined automatically
```

```python
print(np.reshape(a, (-1, 1)))   # -1 means the number of rows will be determined automatically
```

There are two popular ways to flatten an array: .flatten() and .ravel(). The primary difference between the two is that the new array created using ravel() is actually a reference to the parent array (i.e., a “view”). This means that any changes to the new array will affect the parent array as well. Since ravel does not create a copy, it’s memory efficient.

```python
arr = np.array([[1, 2], [3, 4]])
flat1 = arr.flatten()   # returns a copy (independent)
flat2 = arr.ravel()     # returns a view (linked to original)
print("Flatten:", flat1)
print("Ravel:", flat2)
```

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6]])
```

```python
concat1 = np.concatenate((m1, m2), axis=0)  # stack rows
concat2 = np.concatenate((m1, m2.T), axis=1) # stack columns
print("Row concat:\n", concat1)
print("Col concat:\n", concat2)
```

```python
nums = np.array([10, 20, 30, 40, 50])
cond = np.where(nums > 25)    # np.where(condition) → returns an array of indices where the condition is True
print("Condition indices:", cond)
print("Condition values:", nums[cond])  # values > 25
```

## Boxplots

A **Boxplot** shows the distribution of a dataset based on five summary statistics:

- **Minimum**
- **First quartile (Q1 / 25th percentile)**
- **Median (Q2 / 50th percentile)**
- **Third quartile (Q3 / 75th percentile)**
- **Maximum**

It also helps identify **outliers** (points that fall outside the whiskers).


```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```

```python
sns.catplot(x="size", y="total_bill", hue="smoker", kind="boxen", data=tips)
```

### Boxplot Summary (Quick Guide)

- **Median** → the line inside the box, central value of the data.  
- **IQR (Interquartile Range)** → the box height, shows data spread.  
- **Whiskers** → range of typical values (excluding outliers).  
- **Outliers** → points outside whiskers, unusual/extreme values.  
- **Comparison** → compare median, IQR, and outliers across groups.

# Matplotlib

This cheat sheet demonstrates **common plots** used in Machine Learning and EDA with the Seaborn `tips` dataset.

```python
# Line Plot – Trend of Total Bill
plt.plot(tips.index, tips["total_bill"], color="blue", label="Total Bill")
plt.title("Line Plot of Total Bill")
plt.xlabel("Index")
plt.ylabel("Total Bill ($)")
plt.legend()
plt.show()
```

```python
#Purpose: Examine the relationship between total bill and tip.
plt.scatter(tips["total_bill"], tips["tip"], color="red", alpha=0.6)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
#Visualize distribution of total bill; detect skew or unusual values
plt.hist(tips["total_bill"], bins=15, color="green", edgecolor="black")
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Compare average total bill across weekdays
avg_bill = tips.groupby("day")["total_bill"].mean()
plt.bar(avg_bill.index, avg_bill.values, color="purple")
plt.title("Average Total Bill per Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
#Identify outliers and observe data spread per day
plt.boxplot([tips[tips["day"]=="Thur"]["total_bill"],
             tips[tips["day"]=="Fri"]["total_bill"],
             tips[tips["day"]=="Sat"]["total_bill"],
             tips[tips["day"]=="Sun"]["total_bill"]],
            labels=["Thur", "Fri", "Sat", "Sun"])
plt.title("Boxplot of Total Bill per Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

# Comparison: Matplotlib vs Seaborn

| Feature / Aspect             | Matplotlib                             | Seaborn                                      |
|-------------------------------|----------------------------------------|---------------------------------------------|
| **Purpose**                  | General-purpose plotting library       | High-level statistical data visualization  |
| **Ease of Use**               | More control, but verbose syntax       | Easier syntax, less code to produce plots  |
| **Default Styles**            | Basic, plain by default                | Attractive default themes and colors       |
| **Integration**               | Works with any Python data (NumPy, lists, DataFrames) | Works best with pandas DataFrames          |
| **Plot Types**                | Line, scatter, bar, histogram, boxplot, etc. | High-level plots: violin, swarm, pairplot, catplot, heatmap |
| **Customization**             | Highly customizable (axes, ticks, legends) | Customizable, but less low-level control  |
| **Use in ML / EDA**           | Plot trends, learning curves, comparisons | Quick visualizations, explore distributions, categorical data |
| **Learning Curve**            | Moderate to steep                       | Easy for beginners, faster to create nice visuals |
| **Combination**               | Can be combined with Seaborn (Seaborn uses Matplotlib backend) | Often built on top of Matplotlib           |

---

### Key Takeaways

- **Matplotlib** → flexible, full control, backbone of Python plotting.  
- **Seaborn** → high-level, prettier plots, built for statistical and ML data exploration.  
- **Tip:** Use **Seaborn** for quick EDA and **Matplotlib** for custom, detailed plots.
