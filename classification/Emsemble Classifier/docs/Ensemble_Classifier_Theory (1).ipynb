# Ensemble Models in Machine Learning

Think of ensemble models as an orchestra of musicians, where each person specializes in a specific instrument like piano, trumpet, drum, and more. The combination of those skills creates a harmonious melody.

 **Ensemble learning uses the same logic:**

It combines multiple algorithms to obtain better predictive performance than the one from a single model. There is no predefined number of models to consider, and some business goals may require more models than others.


---

## Types of Ensemble Models

In machine learning literature, ensemble methods are generally divided into two main categories:

### 1. Parallel Methods
- Each base learner is trained independently of the others.  
- Training occurs in parallel, so the models do not influence one another.  
- Examples include Bagging and Random Forests.

Parallel methods can be further divided into:
- Homogeneous Ensembles: Use the same base learning algorithm (e.g., multiple decision trees in Random Forest).  
- Heterogeneous Ensembles: Use different algorithms (e.g., combining SVM, Decision Tree, and Logistic Regression).

---

### 2. Sequential Methods
- A new base learner is trained to correct the errors made by the previous model.  
- Models are built sequentially, where each stage depends on the performance of the previous one.  
- Examples include Boosting techniques like AdaBoost, Gradient Boosting, and XGBoost.

##Bagging vs Boosting (Summary)

**Bagging (Bootstrap Aggregating)**  
- Trains multiple models **in parallel** and **independently** on random subsets of the data.  
- Main goal: **Reduce Variance** and prevent **Overfitting**.  
- Famous example: **Random Forest**

**Boosting**  
- Builds models **sequentially**, where each new model **corrects errors** of the previous ones.  
- Main goal: **Reduce Bias** and improve accuracy.  
- Famous examples: **AdaBoost, Gradient Boosting, XGBoost**

## Boosting Algorithms

Boosting algorithms mainly differ in **how they prioritize mispredicted data instances** when creating a new dataset for the next learner. Two prominent boosting methods illustrate this:

### 1. Adaptive Boosting (AdaBoost)
- AdaBoost assigns **weights to model errors**.  
- When creating a new iteration of the dataset for training the next learner, **misclassified samples from the previous learner receive higher weights**.  
- This causes the next learner to focus more on the previously misclassified samples.

### 2. Gradient Boosting
- Gradient Boosting uses **residual errors** from the previous model as the target for the next learner.  
- Instead of weighting misclassified samples, it tries to **minimize the residual error**, gradually closing the gap left by earlier models.

> **Note:**  
> Scikit-learn (`sklearn`) does not provide pre-defined functions for all boosting implementations.  
> However, the open-source library **XGBoost (Extreme Gradient Boosting)** offers Python implementations for gradient boosting.


## Benefits of Ensemble Learning in Machine Learning

Ensemble learning is a versatile approach that can be applied to machine learning models for:

- **Reduction in Overfitting:** Aggregating predictions from multiple models can reduce overfitting that individual complex models might exhibit.  
- **Improved Generalization:** It generalizes better to unseen data by minimizing variance and bias.  
- **Increased Accuracy:** Combining multiple models gives higher predictive accuracy.  
- **Robustness to Noise:** It mitigates the effect of noisy or incorrect data points by averaging predictions from diverse models.  
- **Flexibility:** It works with diverse models, including decision trees, neural networks, and support vector machines, making it highly adaptable.  
- **Bias-Variance Tradeoff:** Techniques like Bagging reduce variance, while Boosting reduces bias, leading to better overall performance.

There are various ensemble learning techniques, each with its own pros and cons.

# Decision Tree and Random Forest

## Decision Tree

### Quick Review
- **Decision Tree** is a `supervised` machine learning model used for both classification and regression tasks.
- It models decisions based on features by splitting the dataset into smaller subsets.
- Each `node` in the tree represents a `feature`, and `edges` represent the `decision` based on a feature's value.
- The goal is to create a model that predicts the value of a target variable by learning `decision rules` from the features.

### How it works
- A decision tree recursively splits the data into two or more homogeneous sets based on the `most significant attribute`. The splitting is guided by various metrics, like `Information Gain (IG)` or `Gini Impurity`.

### Methods of Decision Trees
1. **Information Gain (IG)**
    - Information gain measures the reduction in entropy (uncertainty) after a dataset is split on a feature.
    - Entropy is calculated as:  

    $$
    Entropy(S) = - \sum p(x)log_2p(x)
    $$
    - The higher the information gain, the better the feature is at splitting the data.

2. **Gini Impurity**
    - Measures the likelihood of an incorrect classification if a random sample is classified based on the distribution of labels.
    - Gini impurity is calculated as:
    $$
    Gini(S) = 1 - \sum p(x)^2
    $$
    - Lower Gini values indicate purer splits.

### Advantages
- **Interpretability:** Tree structure is easy to understand.  
- **Flexibility:** Works with both numerical and categorical data.  
- **Minimal Preprocessing:** No need for feature scaling or normalization.

### Disadvantages
- **Sensitive to Noisy Data:** Can easily overfit.  
- **Limited in Modeling Complex Relationships:** May struggle with complex non-linear patterns.

## Random Forest
### Quick Review
- A **Random Forest** is an ensemble method that `combines multiple decision trees` to improve accuracy and reduce overfitting.
- Each tree is trained on `random subsets` of the data and features, and their predictions are `averaged` for better performance.
- It’s robust, reduces variance, and works well for both classification and regression.

### Steps to Create a Random Forest
1. **Bagging**: Randomly sample subsets of the data (with replacement) for each tree.
2. For each tree, randomly select a subset of features at each split.
3. Train a decision tree on each sampled dataset.
4. Repeat steps 1–3 to build multiple decision trees.
5. **Aggregating**: For predictions, aggregate the outputs of all the trees (e.g., majority vote for classification or averaging for regression).
### Advantages
- **High Accuracy:** Combining multiple trees improves predictive performance.  
- **Overfitting Resistant:** Ensemble approach reduces the risk of overfitting.  
- **Feature Importance:** Can indicate the importance of each feature in predictions.

### Disadvantages
- **Computational Complexity:** Requires more computational resources.  
- **Less Interpretable:** Harder to interpret compared to a single decision tree.

## Information Gain and Entropy

**Decision Trees** use metrics like **Entropy** and **Information Gain (IG)** to decide the best feature to split the data.

### 1. Entropy
- Measures the **uncertainty or impurity** in a dataset.  
- Formula:
$$
Entropy(S) = - \sum_{i=1}^{C} p_i \log_2 p_i
$$
where \(C\) is the number of classes and \(p_i\) is the proportion of samples in class \(i\).  
- Lower entropy means the data is more **pure**.

### 2. Information Gain
- Measures the **reduction in entropy** after splitting the dataset on a feature.  
- Formula:
$$
IG(S, A) = Entropy(S) - \sum_{v \in Values(A)} \frac{|S_v|}{|S|} Entropy(S_v)
$$
where \(S_v\) is the subset of \(S\) for which feature \(A\) has value \(v\).  
- Higher IG indicates a **better feature** for splitting the data.
