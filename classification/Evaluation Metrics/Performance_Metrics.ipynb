# Classification Performance Metrics

## 1. Accuracy and Its Limitations
**Accuracy** is one of the simplest and most commonly used metrics for classification problems.  

**Definition:** The ratio of correctly predicted samples to the total number of samples:

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

**Limitation:** Accuracy alone can be misleading, especially in **Imbalanced Datasets**.  
In imbalanced datasets, the **Minority Class** is underrepresented. A model that always predicts the majority class can have high accuracy but poor real-world performance.

**Cancer Diagnosis Example:**  
In a dataset of 1,000 individuals where only 10 have cancer (the positive class),  
a model that always predicts "no cancer" will achieve:

$$
\text{Accuracy} = \frac{990 + 0}{1000} = 99\%
$$

Yet, it completely fails at identifying actual cancer cases.


## 2. Four Basic Outcomes

For better evaluation, a model's prediction results are divided into the following four states:

| Actual / Predicted | Positive | Negative |
|---------------------|----------|----------|
| **Positive**        | True Positive (TP)  | False Negative (FN) |
| **Negative**        | False Positive (FP) | True Negative (TN)  |

- **True Positive (TP):** The model correctly predicted a positive instance.  
- **True Negative (TN):** The model correctly predicted a negative instance.  
- **False Positive (FP):** The model incorrectly predicted a negative instance as positive (**Type I Error**).  
- **False Negative (FN):** The model incorrectly predicted a positive instance as negative (**Type II Error**).  

---

## 3. Primary Metrics (Sensitivity, Specificity, Precision)

**Sensitivity (Recall):** Ability of the system to correctly identify actual positive cases.  

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

**Specificity:** Ability of the system to correctly identify actual negative cases.  

$$
\text{Specificity} = \frac{TN}{TN + FP}
$$

**Precision:** The correctness of the model’s prediction when it predicts positive.  

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

---

## 4. The F1 Score (Harmonic Mean)

The **F1 Score** is a combined metric that balances Precision (P) and Recall (R).  

$$
F1 = \frac{2 \cdot P \cdot R}{P + R}
$$

The harmonic mean is conservative and acts as a smooth minimum, penalizing poor performance in either **P** or **R**, unlike a simple arithmetic mean.  



## 5. Confusion Matrix and Multi-Class Evaluation

**Definition:** A confusion matrix is a table used to evaluate classification models.  
- Rows represent the **Actual Class**.  
- Columns represent the **Predicted Class**.  

The element \( C_{ij} \) shows the number of samples actually in class \( i \), but classified as class \( j \).  

**Metrics per class:**

- **Recall for class i:**
$$
\text{Recall}_i = \frac{C_{ii}}{\sum_j C_{ij}}
$$

- **Precision for class i:**
$$
\text{Precision}_i = \frac{C_{ii}}{\sum_j C_{ji}}
$$

- **Overall Accuracy:**
$$
\text{Accuracy} = \frac{\sum_i C_{ii}}{\sum_i \sum_j C_{ij}}
$$

---

## 6. Averaging Methods

When there is more than one class, two primary methods are used to combine metrics:

| Method          | How It’s Calculated | Key Feature |
|-----------------|---------------------|-------------|
| **Macro**       | Compute metric (e.g., F1) for each class, then take the average. | Treats all classes equally (unweighted). |
| **Micro**       | Aggregate raw outcomes (TP, FP, FN) across all classes, then compute the metric. | Dominated by majority classes. |

## 7. AUC-ROC (Area Under the Receiver Operating Characteristic Curve)

The **AUC-ROC** is a single scalar value that summarizes a binary classification model's performance across all possible classification thresholds, indicating its ability to distinguish between positive and negative classes.  

- A **perfect model** has an AUC of **1.0**.  
- A **random model** has an AUC of **0.5**.  
- A **higher AUC** signifies better discrimination between classes.  
- In diagnostic studies, values **above 0.8** are generally considered clinically useful.  

---

## How the ROC Curve and AUC are Constructed

1. **Binary Classification Model Output:**  
   The model outputs a probability score for each instance, which is then used to predict a class.  

2. **Varying Thresholds:**  
   The classification threshold is adjusted from 0 to 1, changing the balance between **True Positives (TP)** and **False Positives (FP)**.  

3. **Plotting the Curve:**  
   At each threshold, plot:  

   $$
   \text{TPR (Sensitivity)} = \frac{TP}{TP + FN}
   $$

   against  

   $$
   \text{FPR (1 - Specificity)} = \frac{FP}{FP + TN}
   $$

   This produces the **ROC Curve**.  

4. **Calculating the Area:**  
   The **AUC** is the numerical area under the ROC curve, representing the model’s overall performance across all thresholds.  

---

## Interpreting the AUC Score

- **AUC = 1.0** → Perfect classifier  
- **AUC = 0.5** → No better than random guessing  
- **AUC > 0.5** → Better than random chance  
- **Higher AUC** → Stronger ability to distinguish between positive and negative cases  

---

## Key Considerations

- **Class Imbalance:**  
  In imbalanced datasets, the **Precision-Recall Curve (PRC)** and its **Area Under the Curve (AUC-PR)** may provide a more meaningful evaluation than AUC-ROC, since ROC can be less sensitive to imbalance.  

- **Probabilistic Interpretation:**  
  AUC can be interpreted as the **probability that a randomly chosen positive instance will be**
