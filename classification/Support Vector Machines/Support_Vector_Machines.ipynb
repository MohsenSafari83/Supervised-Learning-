# Support Vector Machines (SVM)

A **Support Vector Machine (SVM)** is a powerful, supervised machine learning algorithm primarily used for **classification** (though it can also be adapted for regression).  
Its goal is to find an optimal boundary—called a **hyperplane**—that separates data points into different classes.

The key idea:  
Find the hyperplane that maximizes the distance (**margin**) to the nearest training data points of any class.  
These closest points are called **support vectors**, and they fully define the optimal hyperplane.  
Maximizing the margin helps the model generalize well to unseen data.

---

## How SVM Works

### 1. Hyperplane
- In a classification problem with **N features**, the data points exist in an **N-dimensional space**.  
- The hyperplane is an **(N−1)-dimensional linear subspace** that separates the data into two classes.  
- Example: In 2-D, the hyperplane is a **line**.

### 2. Optimal Margin
- If multiple hyperplanes can separate the classes,  
  SVM selects the one that **maximizes the margin** (distance to the nearest point from each class).  
- This **maximum-margin hyperplane** provides robust separation.

### 3. Support Vectors
- **Support vectors** are the data points lying closest to the hyperplane (on the margin).  
- They are the critical elements of the training set.  
- Removing non-support vector points does **not** change the decision boundary.

### 4. Soft Margin
- Real-world datasets are often **not perfectly linearly separable** (contain noise/outliers).  
- A **soft margin** allows some misclassification.  
- Trade-off: Accepting a few errors → **larger margin** and **better generalization**.

---

## The Kernel Trick (for Non-linear Data)

A fundamental strength of SVM is handling **non-linearly separable data** using the **kernel trick**.

- If data cannot be separated by a straight hyperplane in the original feature space:  
  → The kernel trick **implicitly maps data to a higher-dimensional space** where it becomes linearly separable.  
- The kernel function computes similarities (dot products) **without explicitly transforming** data into that high-dimensional space.  
- This saves huge computational cost.

---

## Common Kernel Functions

### 1. Linear Kernel
- Formula:  
  $$
  K(x_i, x_j) = x_i \cdot x_j
  $$
- Use when data is **linearly separable**.  
- Simple, fast, and effective for **high-dimensional sparse data** (e.g., text classification).

---

### 2. Polynomial Kernel
- Formula:  
 $$
  K(x_i, x_j) = (\gamma \, x_i \cdot x_j + c)^d
  $$
- Allows for **curved and complex decision boundaries**.  
- Suitable when relationships are **polynomial-like**.  
- Common in **computer vision** tasks.

---

### 3. Radial Basis Function (RBF / Gaussian Kernel)
- Formula:  
  $$
  K(x_i, x_j) = \exp\left(-\gamma \, \|x_i - x_j\|^2 \right)
  $$
- The **most popular kernel**.  
- General-purpose: works well when no prior knowledge exists about the data.  
- Maps data into an **infinite-dimensional space**.  
- Great for complex, non-linear decision boundaries.

---

## Summary
- **SVM** → Finds the **optimal separating hyperplane** with maximum margin.  
- **Support vectors** define the boundary.  
- **Soft margin** balances margin size vs. misclassification.  
- **Kernel trick** makes SVM versatile for non-linear problems.  
- Choice of **kernel** (Linear, Polynomial, RBF) depends on data nature.

## Mathematical Formulation of SVM

A **Support Vector Machine (SVM)** constructs a **hyperplane** (or set of hyperplanes) in a high- or infinite-dimensional space.  
This hyperplane can be used for **classification**, **regression**, or other tasks.

---

### Key Idea
- A good separation is achieved by the hyperplane that has the **largest distance (margin)** to the nearest training data points of any class.  
- This distance is called the **functional margin**.  
- Larger margins usually correspond to **lower generalization error**.

---

### Optimization Problem (Hard Margin Case)

Given training data:  
$$
(x_1, y_1), (x_2, y_2), \dots, (x_m, y_m)
$$
where $$ x_i \in \mathbb{R}^n $$ and labels $$ y_i \in \{-1, +1\}$$

We want to find a hyperplane defined by **$$w \cdot x + b = 0$$** such that:

$$
y_i (w \cdot x_i + b) \geq 1 \quad \forall i
$$

---

### Objective Function
The goal is to maximize the **margin**:  
$$
\text{Margin} = \frac{2}{||w||}
$$
Equivalently, we minimize:

$$
\min_{w,b} \frac{1}{2} ||w||^2
$$

subject to:  
$$
y_i (w \cdot x_i + b) \geq 1, \quad \forall i
$$

---

### Soft Margin Extension
For non-linearly separable data (with noise/outliers), **slack variables** $$xi_i \geq 0 $$ are introduced:

$$
y_i (w \cdot x_i + b) \geq 1 - \xi_i
$$

New optimization problem:

$$
\min_{w,b,\xi} \frac{1}{2} ||w||^2 + C \sum_{i=1}^{m} \xi_i
$$

where:
- \(C > 0\) is the **regularization parameter** (controls trade-off between maximizing margin and allowing misclassification).  

---

###  Support Vectors
- The training points that lie **on the margin boundary** are called **support vectors**.  
- Only these points determine the optimal hyperplane.

## Advantages of Support Vector Machines (SVM)

1. **Effective in high-dimensional spaces**  
   Works well even when data has many features.

2. **Effective when dimensions > samples**  
   Performs reliably even if the number of features exceeds the number of samples.

3. **Memory efficient**  
   Uses only a subset of training points (support vectors) in the decision function.

4. **Versatile**  
   - Supports different **kernel functions** (Linear, Polynomial, RBF, etc.).  
   - Allows custom kernel design for specific tasks.  

---

## Disadvantages of Support Vector Machines (SVM)

1. **Risk of overfitting in high dimensions**  
   When features greatly outnumber samples, careful choice of kernel and regularization is essential.

2. **No direct probability estimates**  
   - SVMs don’t provide probabilities natively.  
   - Probabilities must be computed via costly **five-fold cross-validation**.  

---


## Regression using SVM

Support Vector Machines (SVMs), although generally used for **classification**, can also be extended to **regression tasks**.  
This variant is known as **Support Vector Regression (SVR)**.

Unlike traditional linear regression methods that fit a line to minimize the sum of squared errors, **SVR fits a hyperplane in a continuous space** while maximizing the number of points that fall within a margin (also called the “ε-insensitive tube”).  

- **Epsilon (ε):** A hyperparameter that defines the width of the margin.  
- **Support Vectors:** Data points lying outside or on the margin, which determine the position of the regression function.  

### Intuition
An analogy is like constructing a **flyover or bridge** above buildings. The bridge is kept as thin as possible, but it should still provide shade to the maximum number of buildings.  

In other words:  
- **Linear Regression:** Minimizes the sum of distances from all points to the line.  
- **SVR:** Maximizes the number of points within a margin, while keeping the margin width minimal.  



### Advantages of SVR over Normal Regression

1. **Non-linearity**  
   - SVR can capture non-linear relationships using the **kernel trick**.  
   - Linear Regression assumes strict linearity, while non-linear regression is computationally expensive.

2. **Robustness to Outliers**  
   - SVR minimizes errors **within a margin (ε-insensitive zone)**.  
   - Less influenced by extreme outliers compared to Linear Regression.  

3. **Sparsity of Support Vectors**  
   - Only a subset of data points (support vectors) define the model.  
   - More memory-efficient and computationally faster on large datasets.  
   - New data points inside the margin do **not** affect the model.  

4. **Control over Model Complexity**  
   - Hyperparameters (C, ε, kernel parameters) allow fine-tuning of complexity vs. generalization.  
   - Linear Regression does not provide this level of flexibility.  

---

**Key takeaway:**  
SVR is a powerful regression tool when data shows **non-linear patterns, presence of outliers, or high dimensionality**, where simple linear regression falls short.

#Iris Dataset Introduction

The **Iris dataset** is one of the most famous and widely used datasets in machine learning.  
It was introduced by **Ronald Fisher** in 1936 and, due to its simplicity and well-defined structure, is often used as a first example for **multi-class classification** problems.

---

### Key Features of the Dataset
The dataset contains **150 samples**, each belonging to one of three species of the Iris flower.

| Feature            | Data Type         | Description                        |
|--------------------|------------------|------------------------------------|
| Sepal Length       | Numeric (cm)     | Length of the protective leaf      |
| Sepal Width        | Numeric (cm)     | Width of the protective leaf       |
| Petal Length       | Numeric (cm)     | Length of the colored flower part  |
| Petal Width        | Numeric (cm)     | Width of the colored flower part   |

---

### Target Variable
The target variable is the **species of Iris**, which includes three distinct classes:

| Value | Species     | Number of Samples |
|-------|-------------|--------------------|
| 0     | Setosa      | 50                 |
| 1     | Versicolor  | 50                 |
| 2     | Virginica   | 50                 |

---

###  Key Point
The Iris dataset is a **balanced dataset**, since each class has exactly **50 samples**.  
This makes **Accuracy** a reliable metric for evaluating classification models on this dataset.

##1. Setup and Data Loading

```python
# Import essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Convert to DataFrame for easier EDA and visualization
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = [target_names[i] for i in y]
```

##2. Deep Data Preprocessing (EDA) and Visualization

```python
# Pair Plot: Visualize feature relationships and class separation
# This is a powerful EDA tool for multi-dimensional data.
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pair Plot: Feature Relationships by Species", y=1.02)
plt.show()
```

**Interpreting the Pair Plot**

The **pair plot** is one of the most powerful tools for **Exploratory Data Analysis (EDA)**:

- **Main diagonal:** Displays a histogram or kernel density estimate (KDE plot) for each individual feature.  
- **Off-diagonal:** Shows scatter plots that visualize the relationships between every pair of features.  

**Insights on the Iris Datase**:
- The **Setosa species** (often represented in dark blue/purple) is **clearly separable** from the other two species across multiple feature pairs.  
- However, distinguishing between **Versicolor** and **Virginica** is more challenging, as their feature values tend to overlap.  
- This overlap requires a **more complex decision boundary**, which is exactly where algorithms like **SVM with an RBF kernel** excel in creating a non-linear separation between the two species.


```python
# Compute correlation matrix (exclude species column)
corr_matrix = df.drop('species', axis=1).corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    cmap='viridis',
    annot=True,
    fmt=".2f",           # two decimal places
    linewidths=.5,       # grid lines
    cbar_kws={'label': 'Correlation Coefficient'}
)

plt.title('Feature Correlation Heatmap for Iris Dataset', fontsize=14)
plt.show()
```

**Interpretation of the Heatmap**

After running the above code, the heatmap provides insights into the correlations among the Iris dataset features:

- **Strong Positive Correlation (values close to 1.0, usually shown in light/yellow colors):**  
  Indicates that as one feature increases, the other feature also increases proportionally.  
  *Example:* `Petal Length` and `Petal Width` in the Iris dataset show a very high correlation.

- **Weak or No Correlation (values close to 0, usually shown in green/mid-range colors):**  
  Suggests that the features act independently of one another.

- **Effect on the Model:**  
  A very high correlation between two features (like Petal Length and Petal Width) can indicate **multicollinearity**.  
  However, in the case of **SVM**, this is usually less problematic than in linear regression models.  
  Instead, it often points to **redundancy** in the data, meaning that both features carry similar information.


##3. Data Splitting and Feature Scaling

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

#Feature Scaling is mandatory for SVM as it is a distance-based algorithm.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##4. SVM Model Training and Evaluation

```python
# Train the SVM model (C-Support Vector Classification)
# Using the Radial Basis Function (RBF) kernel is standard practice.
# C=1.0 is the regularization parameter.
classifier = SVC(kernel='rbf', random_state=42, C=0.1)

# Fit the model
classifier.fit(X_train_scaled, y_train)

# Make predictions on the scaled test set
y_pred = classifier.predict(X_test_scaled)
```

```python
# Evaluation Metrics
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy:.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```

```python
report = classification_report(y_test, y_pred, target_names=target_names)
print("\nClassification Report:\n", report)
```

## Parameter C: Controlling Regularization in SVM

The **C parameter** in Support Vector Machines is a **regularization parameter** that controls the trade-off between two competing objectives:

1. Achieving a **wider, softer margin**.  
2. Ensuring **fewer misclassifications** on the training data.

---

###  Effect of C on Results

Your observed results (97% vs 100% accuracy) perfectly demonstrate this balance:

#### 1. Small C (e.g., `C=0.1` → Accuracy ≈ 97%)
- **Model Priority:** Maximizing the margin, even if it allows some misclassifications.  
- **Explanation:** With a small value of C, the model is told to “stay soft,” tolerating a few errors (the 3% misclassified points) so that the hyperplane is placed in a more general position.  
- **Goal:** Better generalization to unseen or noisy data.

#### 2. Moderate C (e.g., `C=1` → Accuracy = 100%)
- **Model Priority:** Reducing errors as much as possible, ensuring that all training points are correctly classified if feasible.  
- **Explanation:** With `C=1`, the model receives enough penalty to find the most accurate separating hyperplane. Since the **Iris dataset** is perfectly separable, the model achieves **100% ac**


##5. Decision Boundary Visualization (Using PCA)

```python
# 1. Apply PCA to reduce dimensions to 2
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# 2. Re-train a new SVM classifier using the 2D PCA data
classifier_pca = SVC(kernel='rbf', random_state=42, C=1.0)
classifier_pca.fit(X_train_pca, y_train)

# 3. Create the mesh grid for plotting
# Combine PCA data for a clean plot area
X_set, y_set = np.vstack((X_train_pca, X_test_pca)), np.hstack((y_train, y_test))
X1, X2 = np.meshgrid(np.arange(X_set[:, 0].min() - 1, X_set[:, 0].max() + 1, 0.01),
                     np.arange(X_set[:, 1].min() - 1, X_set[:, 1].max() + 1, 0.01))

# 4. Plot the decision boundary
plt.contourf(X1, X2, classifier_pca.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = plt.cm.RdBu)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
# 5. Plot the actual data points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = [plt.cm.RdBu(i / (len(np.unique(y_set)) - 1))], label = target_names[j], edgecolors='k', s=40)

plt.title('SVM (RBF Kernel) Decision Boundary (Visualized using PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
